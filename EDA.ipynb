# Exploratory Data Analysis (EDA)
## Credit Card Fraud Detection

**Objective:**
To understand the structure, distribution, and imbalance of the data|set
in order to choose appropriate preprocessing, metrics, and models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("data/creditcard.csv")
df.head()
```

The dataset contains anonymized transaction features and a binary target
variable `Class` where 1 indicates fraud.

```python
df.shape
df.info()
df.describe()
```

No missing values are present. Features V1–V28 appear to be PCA-transformed.


```python
class_dist = df['Class'].value_counts()
class_dist
```

```python
class_dist.plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

The dataset is extremely imbalanced with fraudulent transactions forming
less than 1% of the data. Therefore, accuracy is not an appropriate metric.
Recall and precision will be prioritized

```python
df.iloc[:, 1:29].hist(figsize=(16, 12), bins=30)
plt.suptitle("Feature Distributions (V1–V28)")
plt.show()
```

Most features are centered around zero, confirming PCA-like transformations.
Standard scaling will still be applied for linear models.

```python
df['Amount'].describe()
```

```python
df.boxplot(column="Amount", by="Class", figsize=(6,4))
plt.title("Transaction Amount by Class")
plt.suptitle("")
plt.show()
```

Fraudulent transactions tend to have smaller transaction amounts on average,
which may be a useful signal for classification.

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap="coolwarm", vmax=0.6)
plt.title("Feature Correlation Heatmap")
plt.show()
```

No strong correlations with the target variable are visible due to PCA
transformation, supporting the use of machine learning models.

## EDA Summary

- Dataset is highly imbalanced (<1% fraud)
- No missing values
- Features are PCA-transformed
- Amount feature shows different behavior for fraud
- Accuracy is unsuitable; recall and precision are key metrics
